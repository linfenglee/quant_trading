```python
#%%
from importlib import reload
from datetime import datetime
from vnpy.trader.constant import Interval
from vnpy.app.portfolio_strategy.backtesting import BacktestingEngine

# vt_symbol="XBTUSD.BITMEX" 
def run_backtesting(
    strategy_class, 
    setting=None,
    vt_symbols=["btcusdt.BINANCE", "ethusdt.BINANCE"],
    interval=Interval.HOUR,
    start=datetime(2017, 1, 1), 
    end=datetime(2021, 12, 31), 
    rates={"btcusdt.BINANCE": 2/10000, "ethusdt.BINANCE": 2/10000}, 
    slippages={"btcusdt.BINANCE": 0.5, "ethusdt.BINANCE": 0.5}, 
    sizes={"btcusdt.BINANCE": 1, "ethusdt.BINANCE": 1}, 
    priceticks={"btcusdt.BINANCE": 0.5, "ethusdt.BINANCE": 0.5}, 
    capital=1_000_000,
):
    engine = BacktestingEngine()
    
    engine.set_parameters(
        vt_symbols=vt_symbols,
        interval=interval,
        start=start,
        end=end,
        rates=rates,
        slippages=slippages,
        sizes=sizes,
        priceticks=priceticks,
        capital=capital,
    )

    if setting is None:
        setting = {}
    engine.add_strategy(strategy_class, setting)
    
    engine.load_data()
    engine.run_backtesting()
    df = engine.calculate_result()
    engine.calculate_statistics()
    engine.show_chart()
    
    return df

def show_portfolio(df_list):
    portfolio_df = df_list[0]
    for df in df_list[1:]:
        portfolio_df += df
    
    engine = BacktestingEngine()
    engine.calculate_statistics(portfolio_df)
    engine.show_chart(portfolio_df)
```
